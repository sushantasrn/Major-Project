```python
inputs=int(input("ENTER NO OF NODE:"))
graph=[]
graph.append([[0,inputs],inputs])

for r in range(1,inputs):
    graph.append([[0,inputs],r])
#print(graph)
#print(graph[6][1])

queue=[]
queue.append(inputs)
#queue.append(0)
pair=[]
pop_list=[]
while queue!=[]:
    i=queue[0]
    
    if i%2==0:

        l=0
        
        for j in range(len(graph)):
            if graph[j][1]==i:
                
                l=graph[j][0]              
              
                break

        l=list(set(l))
        
            
        for j in range(len(l)):

            val=(l[j]+i)//2

            if val!=i:
               # print(l[j])
                if l[j]==0:
                    #print(pair)
                    pair.append([i,val,1])
                if l[j]==inputs:
                    pair.append([i,val,2])
                if l[j]!=0 and l[j]!=inputs:
                    pair.append([i,val,0])
                if val not in queue and val not in pop_list:
                    queue.append(val)
                    
                graph[val][0].append(i)
                graph[val][0].extend(graph[i%inputs][0])
                
                
        pop_list.append(queue[0])
        queue.remove(queue[0])
        

        
    else:
        pop_list.append(queue[0])
        queue.remove(queue[0])
        
    
print("pair:",pair)  

    
    
    
```

```python
arrr=[[0]*inputs]*inputs
for i in range (0,inputs):
    for j in range(0,inputs):
        arrr[i][j]=-1

for i in range(len(pair)):    
    l1=pair[i]
    s=l1[0]
    d=l1[1]
    c=l1[2]
    arrr[s-1][d-1]=c
    print(s,d,c)
print(arrr)
```
